## Scatter Plot - Applying Colour

The goal of this section is to demonstrate how colour can be applied to a simple scatter plot.  
We start with:
1️. Default colours  
2️. A single chosen named colour (e.g., "blue")  
3️. Another example using a different colour name (e.g., "coral")

This aligns with the first colour-modification steps from Workshop 3.

```python
import matplotlib.pyplot as plt

# Provided sample data
x = [6, 7, 8, 7, 2, 3, 9, 4, 10, 15, 9, 7]
y = [98, 85, 87, 88, 112, 104, 87, 94, 78, 78, 85, 86]

# 1️ Basic scatter (default color)
plt.figure()
plt.scatter(x, y)
plt.title("Basic Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()

# 2️ Scatter with blue color
plt.figure()
plt.scatter(x, y, color="blue")
plt.title("Scatter with Blue Points")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()

# 3️ Scatter with coral color
plt.figure()
plt.scatter(x, y, color="coral")
plt.title("Scatter with Coral Points")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

## Marker Shapes, Edge Colours, and Fill Colours

In this visualization, we apply more colour control to scatter markers:
- `marker` changes the point symbol (e.g., `"o"`, `"s"`, `"D"`, `"^"`)
- `edgecolors` defines the outline colour
- `facecolors` defines the inside fill colour
- `s` controls marker size

This demonstrates finer colour styling as required in Workshop 3.


```python
# Square markers with blue edge and yellow fill
plt.figure()
plt.scatter(x, y, marker="s", s=120, edgecolors="blue", facecolors="yellow")
plt.title("Squares - Blue Edge, Yellow Fill")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()

# Diamond markers with black edge and coral fill
plt.figure()
plt.scatter(x, y, marker="D", s=120, edgecolors="black", facecolors="coral")
plt.title("Diamonds - Black Edge, Coral Fill")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()

# Circle markers with thick edge and transparent fill
plt.figure()
plt.scatter(x, y, marker="o", s=200, edgecolors="purple", facecolors="none", linewidth=2)
plt.title("Circles - Purple Edge, No Fill")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

## Histogram with Colour, Edge and Transparency (Alpha)

This section demonstrates colour styling in histograms:
- `color` or `facecolor` → fill colour
- `edgecolor` → bar boundary
- `alpha` → transparency (0 = transparent, 1 = solid)

We will also add `bins` to control bar detail.

This satisfies the histogram styling requirement of Workshop 3. :contentReference[oaicite:0]{index=0}

```python
import seaborn as sns

# Load iris dataset (if not already loaded)
iris = sns.load_dataset('iris')

# Histogram of petal_length with styling
plt.figure()
plt.hist(
    iris['petal_length'],
    bins=12,                  # number of bars
    color="skyblue",          # fill colour
    edgecolor="black",        # border colour
    alpha=0.75                # transparency
)
plt.title("Histogram of Petal Length - Styled")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()

# Variation: coral fill + thicker edges
plt.figure()
plt.hist(
    iris['petal_length'],
    bins=12,
    color="coral",
    edgecolor="darkred",
    linewidth=1.5,
    alpha=0.6
)
plt.title("Histogram of Petal Length - Coral Theme")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

## Multi-Line Plot with Colour-Coded Classes

To show how colour helps distinguish categories (classes),
we will plot three lines - one for each species in the Iris dataset.

Colour roles:
- Represent species grouping
- Aid visual comparison
- Automatically generate a legend

This fulfills the “multi-line coloured plot” requirement of Workshop 3.

```python
# Group by species and plot each separately with colour
plt.figure()

for species, group in iris.groupby('species'):
    plt.plot(
        group.index,
        group['sepal_length'],
        label=species.capitalize()
    )

plt.title("Multi-Line Plot - Sepal Length by Species")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.legend(title="Species")
plt.show()
```
